### Libraries

```python
import numpy as np
from nitime.timeseries import TimeSeries
from nitime.analysis import SpectralAnalyzer
import os
import matplotlib.pyplot as plt
```

```python
def repetition_time(filename):

    site = filename.split('/')[-1].split('_')[-2]

    if 'Amsterdam-AMC' in site:
        TR = 2.375
    elif 'Amsterdam-VUmc' in site:
        TR = 1.8
    elif 'Barcelona-HCPB' in site:
        TR = 2
    elif 'Bergen' in site:
        TR = 1.8
    elif 'Braga-UMinho-Braga-1.5T' in site:
        TR = 2
    elif 'Braga-UMinho-Braga-1.5T-act' in site:
        TR = 2
    elif 'Braga-UMinho-Braga-3T' in site:
        TR = 1
    elif 'Brazil' in site:
        TR = 2
    elif 'Cape-Town-UCT-Allegra' in site:
        TR = 1.6
    elif 'Cape-Town-UCT-Skyra' in site:
        TR = 1.73
    elif 'Chiba-CHB' in site:
        TR = 2.3
    elif 'Chiba-CHBC' in site:
        TR = 2.3 
    elif 'Chiba-CHBSRPB' in site:
        TR = 2.5 
    elif 'Dresden' in site:
        TR = 0.8 
    elif 'Kyoto-KPU-Kyoto1.5T' in site:
        TR = 2.411 
    elif 'Kyoto-KPU-Kyoto3T' in site:
        TR = 2
    elif 'Kyushu' in site:
        TR = 2.5
    elif 'Milan-HSR' in site:
        TR = 2
    elif 'New-York' in site:
        TR = 1
    elif 'NYSPI-Columbia-Adults' in site:
        TR = 0.85
    elif 'NYSPI-Columbia-Pediatric' in site:
        TR = 0.85
    elif 'Yale-Pittinger-HCP-Prisma' in site:
        TR = 0.8
    elif 'Yale-Pittinger-HCP-Trio' in site:
        TR = 0.7
    elif 'Yale-Pittinger-Yale-2014' in site:
        TR = 2
    elif 'Bangalore-NIMHANS' in site:
        TR = 2 
    elif 'Barcelone-Bellvitge-ANTIGA-1.5T' in site:
        TR = 2
    elif 'Barcelone-Bellvitge-COMPULSE-3T' in site:
        TR = 2
    elif 'Barcelone-Bellvitge-PROV-1.5T' in site:
        TR = 2
    elif 'Barcelone-Bellvitge-RESP-CBT-3T' in site:
        TR = 2
    elif 'Seoul-SNU' in site:
        TR = 3.5
    elif 'Shanghai-SMCH' in site:
        TR = 3
    elif 'UCLA' in site:
        TR = 2
    elif 'Vancouver-BCCHR' in site:
        TR = 2
    elif 'Yale-Gruner' in site:
        TR = 2
    else:
        raise ValueError(f"Site '{site}' does not have a defined TR value in TR_mappings. Please add it.")
        
    # print(f"Site: {site}")
    
    return TR
```

### Example run

```python
# Set the path to the file
file_path = '/pscratch/sd/p/pakmasha/MBBN_data/Brazil_sub-P00196820141124/Brazil_sub-P00196820141124.npy'

# Ensure the file exists
if not os.path.exists(file_path):
    raise FileNotFoundError(f"File not found at {file_path}")

# Load the resting-state fMRI time series data
data = np.load(file_path)
data = data.T

print(f"Data shape: {data.shape}")

# Parameters
TR = 2.0  # Replace this with your actual TR (Repetition Time) in seconds

# Initialize list to store maximum frequencies per ROI
max_frequencies = []

# Loop through each ROI to compute PSD and find maximum frequency
for roi_index in range(data.shape[0]):
    roi_time_series = data[roi_index, :]

    # Create TimeSeries object
    T = TimeSeries(roi_time_series, sampling_interval=TR)

    # Perform spectral analysis
    S = SpectralAnalyzer(T)

    # Extract frequency and power data
    frequencies = np.array(S.spectrum_fourier[0][1:])  # Skip the first frequency (DC component)
    power = np.abs(S.spectrum_fourier[1][1:])          # Corresponding power values

    # Find the maximum frequency
    max_frequency = np.max(frequencies)
    max_frequencies.append(max_frequency)

    if roi_index < 10:
        # Plot the PSD for the current ROI
        plt.figure(figsize=(8, 4))
        plt.plot(frequencies, power, label=f'ROI {roi_index}')
        plt.xlabel('Frequency (Hz)')
        plt.ylabel('Power')
        plt.title(f'Power Spectral Density for ROI {roi_index}')
        plt.legend()
        plt.grid(True)
        plt.show()
    
        plt.figure(figsize=(8, 4))
        plt.plot(np.log10(frequencies), np.log10(power), label=f'ROI {roi_index}')
        plt.xlabel('Frequency (Hz)')
        plt.ylabel('Power')
        plt.title(f'Log Power Spectral Density for ROI {roi_index}')
        plt.legend()
        plt.grid(True)
        plt.show()

# Find the maximum frequency value across all ROIs
overall_max_frequency = max(max_frequencies)

# Print the results
print(f"Maximum frequency for each ROI: {max_frequencies}")
print(f"Overall maximum frequency: {overall_max_frequency}")
```

### Apply the is_filtered function to all time-series data

```python
base_path = "/pscratch/sd/p/pakmasha/MBBN_data"
count = 0

# Walk through all subdirectories and process .npy files
for root, dirs, files in os.walk(base_path):
    for file in files:
        if file.endswith('.npy') and file.split('_')[-1] != "0.1.npy":
            filename = os.path.join(root, file)
            is_filtered(filename)
            count += 1

print(f"Total of {count} files processed.")
```
