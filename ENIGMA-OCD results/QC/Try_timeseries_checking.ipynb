```python
import pandas as pd

# Replace 'path_to_your_file' with the actual file path
file_path = '/pscratch/sd/p/pakmasha/ENIGMA unzip/Amsterdam_AMC/halfpipe/sub-101/func/sub-101_task-rsfMRI_run-3_feature-corrMatrix2_atlas-schaefer2011CombinedDseg_timeseries.tsv'
data = pd.read_csv(file_path, sep='\t', header=None)
```

```python
# Count NaNs in each column
nan_counts = data.isna().sum()

# Filter columns with NaNs
nan_columns = []  # Empty list to store column numbers with NaNs
for col_num, nan_count in nan_counts.items():
    if nan_count > 0:
        nan_columns.append(col_num)  # Append column number with NaNs
        print(f"Column {col_num}: {nan_count} NaN values")

# Print the list of column numbers with NaNs
print(f"Columns with NaNs: {nan_columns}")
print(f"Total {len(nan_columns)} columns with NaNs")
```
